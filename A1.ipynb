**Assignment on Linear Regression:**
The  following  table  shows  the  results  of  a recently  conducted  study  on  the correlation of the number of hours spent driving with the risk of developing acute backache. Find the equation of the best fit line for this data.
a1.PNG

```
from statistics import mean
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

style.use('fivethirtyeight')
```

```
xs = np.array([10, 9, 2, 15, 10, 16, 11, 16])
ys = np.array([95, 80, 10, 50, 45, 98, 38, 93])
plt.scatter(xs , ys)
```

```
def best_fit_slope_intercept(xs, ys):
    m = ((mean(xs) * mean(ys)) - mean(xs*ys)) / (mean(xs)*mean(xs) - mean(xs*xs))
    b = mean(ys) - m*mean(xs)
    return m, b
```

```
m,b = best_fit_slope_intercept(xs,ys)
print('Slope = ' , m)
print('Intercept = ', b)
```

```
regression_line = [(m*x)+b for x in xs]
```

```
predict_x = 12
predict_y = (m*predict_x) + b

print(predict_y)
```

```
plt.scatter(xs, ys)
plt.plot(xs, regression_line , color='g')
plt.scatter(predict_x, predict_y, color='r' , marker='*' , s=300)
plt.show()
```
